```python
#importing  libraries
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt=pd.read_csv(r"C:\Users\Amol purkar\Desktop\Imarticus\DATA SETS\snsdata.csv")
```

```python
dt
```

```python
# data cleaning
dt.isnull().sum()[dt.isnull().sum()>0]
```

```python
dt.age.describe()
```

```python
dt.gender=dt.gender.fillna("F")
dt.age=dt.age.fillna(dt.age.mean())
```

```python
dt.gender=dt.gender.replace({"M":0,"F":1})
```

```python
#model building
ssd=[]

for k in range(2,12):
    kmeans_sns=KMeans(n_clusters=k)
    kmeans_sns.fit(dt)
    ssd.append(kmeans_sns.score(dt))
    
```

```python
ssd=(np.abs(ssd))
ssd
```

```python
ssd=np.round(ssd)
```

```python
cluster_value=list(range(2,12))
```

```python
plt.plot(cluster_value,ssd, marker="*")
plt.grid()
```
